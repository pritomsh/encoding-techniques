#  Data Separation

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size=.30,random_state=1)
```

```python
xtrain.shape
```

```python
xtest.shape
```

#  Apply Decision Tree Algorithm 

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
#make object 
dt_object = DecisionTreeClassifier(random_state=5)
```

```python
#fit model
dt_object.fit(xtrain,ytrain)
```

```python
#predict ytest 
pred = dt_object.predict(xtest)
```

```python
pred
```

#  Confusion Matrix 

```python
from sklearn.metrics import classification_report, accuracy_score, confusion_matrix, plot_roc_curve,roc_curve
```

```python
print(classification_report(ytest, pred))
```

```python
print(classification_report(ytest, pred))
```

```python
plt.figure(figsize=(5,3))
sns.heatmap(confusion_matrix(ytest, pred), annot=True, fmt='d',
            xticklabels=["Positive", "Negative"], 
            yticklabels=["Positive", "Negative"])
plt.title('Confusion Matrix')
plt.show()
```

```python
accuracy_score(ytest, pred)
```

```python
tp, fn, fp, tn =  confusion_matrix(ytest, pred).reshape(-1)
```

```python
print(" True positive :{}\n True negative :{} \n False positive :{} \n False negative  :{} \n ".format(tp,tn, fp,fn ))
```

```python
print("Total Event = ", tp + fn + fp + tn)
```

```python
# Accuracy
# Accuracy = (TP + TN ) / (TP + FP + TN + FN)
print('Accuracy = ' ,(tp+tn)/(tp + fn + fp + tn))
```

```python
# Precision or Positive Predictive Value (PPV)
# PPV = True Positive / (True Positive + False Positive)
precision = tp/( tp + fp)
print("Precision = ",precision ) 
```

```python
# Sensitivity, Recall, Hit Rate, or True Positive Rate (TPR)
# TPR = True Positive / (True Positive + False Negative)
recall = tp/(tp+fn)
print("True Positive Rate = ",recall )
```

```python
# False Positive Rate (FPR) or 1-Specificity
# FPR = FP / (FP + TN)
print("False Positive Rate = ", fp/(fp+tn))
```

```python
# F1 Measure
# F1 Measure = (Precision + Recall) / 2
print("F1 Measure = ", (precision+recall)/2)
```

```python
# Harmonic Mean, F1 Score
# F1 = ( 2 * Precision * Recall ) / (Precision + Recall)
print("F1 Score = ", (2*precision*recall)/(precision+recall))
```

```python
# Specificity, Selectivity or True Negative Rate (TNR)
# Specificity = True Negative / (True Negative + False Positive)

print("Specificity = ", tn/(tn+fp))
```

```python
# Threat Score (TS) or Critical Success Index (CSI)
# CSI = TP / (TP + FN + FP )
print("Critical Success Index = ", tp/(tp+fn+fp))
```

```python
# False Discovery Rate (FDR)
# FDR = FP / (TP + FP )
print("False Discovery Rate = ", fp/(tp+fp))
```

#  The Area Under the Curve


An AUC of 0.5 suggests no discrimination , 0.7-0.8 is considered acceptable, 0.8-0.9 is considered excellent, and more than 0.9 is  considered outstanding.

```python
plot_roc_curve(dt_object, xtest, ytest)
plt.plot([0,1],[0,1])
```
